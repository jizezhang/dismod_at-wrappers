```python
import sys
sys.path.insert(0, '../code/')
import pandas as pd
import covariate
import numpy as np
```

```python
measurement = pd.read_csv('./data/copd/copd_sincidence.csv', low_memory=False)[['location_id', 'location_name' ,'sex', 'age_start', 'age_end',
        'year_start', 'year_end', 'mean', 'standard_error', 'measure']]
elevation = pd.read_csv('./data/copd/elevation.csv')[['location_id', 'sex_id', 'age_group_id', 'year_id', 'mean_value']]
ldi = pd.read_csv('./data/copd/ldi.csv')[['location_id', 'sex_id', 'age_group_id', 'year_id', 'mean_value']]
sev = pd.read_csv('./data/copd/sev_copd.csv')[['location_id', 'sex_id', 'age_group_id', 'year_id', 'mean_value']]
population = pd.read_csv('./data/pop.csv')[['location_id', 'sex_id', 'age_group_id', 'year_id', 'population']]
measurement.dropna(inplace=True)
sev['age_group_id'] = 22
```

```python
elevation.dropna(inplace=True)
ldi.dropna(inplace=True)
sev.dropna(inplace=True)
population.dropna(inplace=True)
measurement = measurement.reset_index(drop=True)
```
